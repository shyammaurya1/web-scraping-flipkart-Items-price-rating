```python
import urllib.request as url
from bs4 import BeautifulSoup
import pandas as pd
#link="https://www.flipkart.com/search?q=samsung&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
link="https://www.flipkart.com/search?q=samsung&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&sort=price_asc"
page=url.urlopen(link)
print(page)
```

```python
soup=BeautifulSoup(page,"html")
```

```python
name=soup.find_all("div",class_="_3wU53n")
price=soup.find_all("div",class_="_1vC4OE _2rQ-NK")
rating=soup.find_all("div",class_='hGSR34')
```

```python
na=[]
pr=[]
ra=[]

for _ in name:
    na.append(_.getText())

for _ in price:
    pr.append(_.getText())

for _ in rating:
    ra.append(_.getText())        
```

```python
t=list(zip(na,pr,ra))
df=pd.DataFrame(t)
df
```
